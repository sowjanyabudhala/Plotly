```python
import pandas as pd
import datashader as ds
import datashader.transfer_functions as tf
from colorcet import fire,kbc,bmw,m_gray
import plotly.express as px
```

```python
df2 = pd.read_csv('/content/Crimes_-_2001_to_Present.csv', on_bad_lines='skip')
```

```python
df2.head(3)
```

```python
#filter the dataframe
df2=df2[['Primary Type','Latitude','Longitude']]
print(df2['Primary Type'].unique())
```

```python
df2=df2[df2['Primary Type'].isin(['PROSTITUTION'])]
df2.dropna(subset=['Latitude','Longitude'],inplace=True)
print(df2.shape)
```

```python
df2['Longitude'] = pd.to_numeric(df2['Longitude'], errors='coerce')
df2['Latitude'] = pd.to_numeric(df2['Latitude'], errors='coerce')

# Drop rows with missing values in Longitude and Latitude
df2.dropna(subset=['Latitude', 'Longitude'], inplace=True)

#build an abstract canvas representing the space in which to plot data

cvs = ds.Canvas(plot_width=500, plot_height=300)
```

```python
#build an abstract canvas representing the space in which to plot data
cvs=ds.Canvas(plot_width=500,plot_height=300)
```

```python
#project the latitude and longitude on to the Canvas and map the data to pixels as points
aggs=cvs.points(df2,x='Longitude',y='Latitude')
```

```python
#aggs is an xarray
coords_lat,coords_lon=aggs.coords['Latitude'].values,aggs.coords['Longitude'].values
```

```python
#set the corners of the image that need to be  passed on the mapbox
coordinates=[[coords_lon[0],coords_lat[0]],
             [coords_lon[-1],coords_lat[0]],
            [coords_lon[-1],coords_lat[-1]],
             [coords_lon[0],coords_lat[-1]]]
```

```python
#set the image color and legend (how) types
#linear(how=linear),logarithmic(how=log),percentile(how=eq_hist)
img=tf.shade(aggs,cmap=fire,how='eq_hist')[::-1].to_pil()
```

```python
#create a mapbox figure with plotly
fig=px.scatter_mapbox(df2[:1],lat='Latitude',lon='Longitude',zoom=10)
```

```python
#Add datashader image as mapbox layer image
fig.update_layout(mapbox_style='carto-darkmatter',mapbox_layers=[{'sourcetype':'image','source':img,'coordinates':coordinates}])
fig.show()
```

**Sunburst treemap**

```python
import numpy as np
```

```python
s=pd.read_csv('/content/MPVDataset.csv')
```

```python
s.head()
```

```python
s['Victim\'s age']=pd.to_numeric(s['Victim\'s age'],errors='coerce').fillna(0).astype(np.int64)
```

```python
s.rename(columns={'Fleeing (Source: WaPo)': 'Fleeing'}, inplace=True)
```

```python
s = s[s['State'].isin(['NY','CA','TX'])]
```

```python
s = s[s["Victim's race"].isin(['White','Black','Hispanic','Asian'])]
```

```python
fig=px.sunburst(s,path=['Unarmed','State',"Victim's race"],#root,branch,leaves
                color='Unarmed',color_discrete_sequence=px.colors.qualitative.Pastel,maxdepth=-1)
fig.show()
```

```python
fig=px.sunburst(s,path=['Fleeing','State',"Victim's race"],#root,branch,leaves
                color='Fleeing',color_discrete_sequence=px.colors.qualitative.Pastel,maxdepth=-1)
fig.update_traces(textinfo='label+percent entry')
fig.update_layout(margin=dict(t=0,l=0,r=0,b=0))
fig.show()
```

```python
fig=px.sunburst(s,path=['Victim\'s gender','State',"Victim\'s race"],#root,branch,leaves
                color='Victim\'s gender',color_discrete_sequence=px.colors.qualitative.Pastel,maxdepth=-1)
fig.update_traces(textinfo='label+percent entry')
fig.update_layout(margin=dict(t=0,l=0,r=0,b=0))
fig.show()
```

```python
fig=px.sunburst(s,path=['Unarmed','State',"Victim's race","Victim's gender"],#1root,2branch,1leaves
                color='Unarmed',color_discrete_sequence=px.colors.qualitative.Pastel,maxdepth=-1) #render all levels in the hierarchy)
fig.update_traces(textinfo='label+percent entry')
fig.update_layout(margin=dict(t=0,l=0,r=0,b=0))
fig.show()
```

```python
fig=px.sunburst(s,path=['Unarmed','State',"Victim's race"],maxdepth=-1,color='Victim\'s age',color_continuous_scale=px.colors.sequential.BuGn,range_color=[10,100]) #render all levels in the hierarchy)
fig.update_traces(textinfo='label+percent entry')
fig.update_layout(margin=dict(t=0,l=0,r=0,b=0))
fig.show()
```

```python
fig=px.sunburst(s,path=['Unarmed','State',"Victim's race"],color_discrete_sequence=px.colors.qualitative.Pastel,branchvalues='remainder')
fig.update_traces(textinfo='label+percent entry')
fig.update_layout(margin=dict(t=0,l=0,r=0,b=0))
fig.show()
```
