```python
import plotly.graph_objects as go
```

```python
# Use for animation rotation at the end
x_eye = -1.25
y_eye = 2
z_eye = 0.5
fig = px.scatter_3d(df, x='GDP per capita', y='% Econ. active', z='Years in school (avg)', color='Continent',
                   color_discrete_sequence=['magenta', 'green'], log_x=True,template='gridon',
                   title='Female labor force participation analysis',
                   labels={'Years in school (avg)': 'Years women are in school'},hover_name='Entity', height=700,)
 #Use for animation rotation
fig.update_layout(scene_camera_eye=dict(x=x_eye, y=y_eye, z=z_eye),
                  updatemenus=[dict(type='buttons',
                                    showactive=False,
                                    y=1,
                                     x=0.8,
                                     xanchor='left',
                                     yanchor='bottom',
                                     pad=dict(t=45, r=10),
                                     buttons=[dict(label='Play',
                                                   method='animate',
                                                   args=[None, dict(frame=dict(duration=250, redraw=True),
                                                                    transition=dict(duration=0),
                                                                    fromcurrent=True,
                                                                    mode='immediate'
                                                                    )]
                                                   )
                                              ]
                                     )
                                ]
                   )


def rotate_z(x, y, z, theta):
    w = x + 1j * y
    return np.real(np.exp(1j * theta) * w), np.imag(np.exp(1j * theta) * w), z


frames = []

for t in np.arange(0, 6.26, 0.1):
     xe, ye, ze = rotate_z(x_eye, y_eye, z_eye, -t)
     frames.append(go.Frame(layout=dict(scene_camera_eye=dict(x=xe, y=ye, z=ze))))
fig.frames = frames

fig.write_html("animation.html")
pio.show(fig)

```
